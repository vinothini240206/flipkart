```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
result=classifier("I love to learn about Machine Learning.")
```

```python
result
```

```python
classifier = pipeline("zero-shot-classification")
result=classifier("This is a course about the Transformers library",candidate_labels=["education","politics","business"],)
```

```python
result
```

```python
generator = pipeline("text-generation", model="distilgpt2")
results=generator(
                  "In this course,we will teach you how to",
                  max_length=30,
                  num_return_sequences=2,)
```

```python
for item in results:
  print(item["generated_text"])
```

```python
ner = pipeline("ner", grouped_entities=True)
result=ner("My name is Santhosh and I work at Guvi Geek Networks in Chennai.")
```

```python
for item in result:
  print(item)
```
